```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train.csv')
df_test=pd.read_csv('test.csv')
```

```python
df_test
```

```python
df.drop('GarageYrBlt',inplace=True,axis=1)
```

```python
for i in df:
    #print(str(df[i].isnull().sum())+"--"+i)
    if df[i].isnull().sum()!=0:
        print(i+'--'+str(df[i].isnull().sum()))
```

```python
df['MiscFeature'].fillna(0,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
for i in df:
    if len(df[i].value_counts())<=25:
        df[i]=encoder.fit_transform(df[i].astype(str))
        df_test[i]=encoder.fit_transform(df_test[i].astype(str))
```

```python
df.loc[:, 'MSSubClass':'SaleCondition']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( df.loc[:, 'MSSubClass':'SaleCondition'], df.SalePrice, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet, SGDRegressor
from sklearn.ensemble import RandomForestRegressor
reg = LinearRegression().fit(X_train, y_train)
reg1 = Lasso().fit(X_train, y_train)
reg2 = Ridge().fit(X_train, y_train)
reg3 = ElasticNet().fit(X_train, y_train)
reg4 = RandomForestRegressor().fit(X_train, y_train)
```

```python
reg4.score(X_train,y_train)
```

```python

from sklearn.metrics import mean_squared_error, r2_score
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(y_test, reg4.predict(X_test)))
```

```python
df_test.drop(['Id'],inplace=True,axis=1)
```

```python
reg4.predict(df_test)
```

```python
df_test.fillna(df_test.mean(), inplace=True)
```
